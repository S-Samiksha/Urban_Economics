### Adding manhattan distance 

This is the absolute distance in the x axis and y axis:


```python
# adapted from: https://kanoki.org/2019/12/27/how-to-calculate-distance-in-python-and-pandas-using-scipy-spatial-and-distance-functions/
def manhattan(lon1, lat1, lon2, lat2):

    lon1, lat1, lon2, lat2 = lon1, lat1, lon2, lat2

    newlon = (lon2 - lon1)*111000
    newlat = (lat2 - lat1)*111000
    
    distance = abs(newlon) + abs(newlat)

    
    return distance


#lat and long from raffles place mrt station 
combined['distance'] = combined.apply(lambda row: manhattan(row.lng_y, row.lat_y, 103.8513,1.2830 ), axis=1)
```

### Adding distance from API

```python
import requests
import json# call the OSMR API

i=1

def router(lon_1,lat_1,lon_2,lat_2):
    global i
    r = requests.get(f"http://router.project-osrm.org/route/v1/car/{lon_1},{lat_1};{lon_2},{lat_2}?overview=false""")# then you load the response using the json libray
    # by default you get only one alternative so you access 0-th element of the `routes`
    
    routes = json.loads(r.content)
    # print(routes)
    route_1 = routes.get("routes")[0]
    # print(f"Overall, the route data is:\n{route_1}\n")
    seconds = route_1['legs'][0]['duration']
    #print(f"Duration: {seconds}sec ({seconds/60:.2f} min )")
    print("row", i)
    print(route_1['legs'][0]['distance'])
    i += 1
    return float(route_1['legs'][0]['distance'])


j=1

def router2(lon_1,lat_1,lon_2,lat_2):
    global j
    r = requests.get(f"http://router.project-osrm.org/route/v1/car/{lon_1},{lat_1};{lon_2},{lat_2}?overview=false""")# then you load the response using the json libray
    # by default you get only one alternative so you access 0-th element of the `routes`
    
    routes = json.loads(r.content)
    # print(routes)
    route_1 = routes.get("routes")[0]
    # print(f"Overall, the route data is:\n{route_1}\n")
    seconds = route_1['legs'][0]['duration']
    #print(f"Duration: {seconds}sec ({seconds/60:.2f} min )")
    print("row", j)
    print(float(seconds/60.0))
    j += 1
    return float(seconds/60.0)
```

```python
combined['travel_distance'] = combined.apply(lambda row: router(row.lng_y, row.lat_y, 103.8513,1.2830), axis=1)
combined['travel_time'] = combined.apply(lambda row: router2(row.lng_y, row.lat_y, 103.8513,1.2830), axis=1)
```

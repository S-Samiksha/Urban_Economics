```python
# https://kanoki.org/2019/12/27/how-to-calculate-distance-in-python-and-pandas-using-scipy-spatial-and-distance-functions/

def manhattan(lon1, lat1, lon2, lat2):

    lon1, lat1, lon2, lat2 = lon1, lat1, lon2, lat2

    newlon = (lon2 - lon1)*111000
    newlat = (lat2 - lat1)*111000
    
    distance = abs(newlon) + abs(newlat)

    
    return distance
```

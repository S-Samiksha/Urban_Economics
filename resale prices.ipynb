```python
llist = ["page"+str(x) for x in range(1,16)]
llist.insert(0, 'combined_dist stable')
print(llist)
combined = pd.read_excel("./combined_dist_stable.xlsx", sheet_name=llist, engine='openpyxl') #import fresh data  
```

```python
final = pd.DataFrame()
for item in llist:
    temp = combined.get(item)
    final = pd.concat([final,temp])
```

```python
import requests
import json# call the OSMR API

i=1


def router(row):
    lon_1, lat_1, lon_2, lat_2 = row['lng_hdb'], row['lat_hdb'], row['lng_mrt'], row['lat_mrt']
    
    global i
    r = requests.get(f"http://router.project-osrm.org/route/v1/car/{lon_1},{lat_1};{lon_2},{lat_2}?overview=false""")# then you load the response using the json libray
    # by default you get only one alternative so you access 0-th element of the `routes`
    
    routes = json.loads(r.content)
    # print(routes)
    route_1 = routes.get("routes")[0]
    # print(f"Overall, the route data is:\n{route_1}\n")
    seconds = route_1['legs'][0]['duration']
    #print(f"Duration: {seconds}sec ({seconds/60:.2f} min )")
    print("row", i)
    print(route_1['legs'][0]['distance'])
    i += 1

    return float(route_1['legs'][0]['distance']), float(seconds/60.0)
    
def walker(row):
    lon_1, lat_1, lon_2, lat_2 = row['lng_hdb'], row['lat_hdb'], row['lng_mrt'], row['lat_mrt']
    
    global i
    r = requests.get(f"http://router.project-osrm.org/route/v1/foot/{lon_1},{lat_1};{lon_2},{lat_2}?overview=false""")# then you load the response using the json libray
    # by default you get only one alternative so you access 0-th element of the `routes`
    
    routes = json.loads(r.content)
    # print(routes)
    route_1 = routes.get("routes")[0]
    # print(f"Overall, the route data is:\n{route_1}\n")
    seconds = route_1['legs'][0]['duration']
    #print(f"Duration: {seconds}sec ({seconds/60:.2f} min )")
    print("row", i)
    print(route_1['legs'][0]['distance'])
    i += 1

    return float(route_1['legs'][0]['distance']), float(seconds/60.0)
    
```

```python
#import os 
#combined.to_csv('./combined_resale.csv') 
shortened = combined.drop_duplicates(subset=['postal'])
shortened.shape
geylangb = shortened[shortened['name']=='GEYLANG BAHRU MRT STATION']
```

```python
# Running by sections - Not used for HDB to MRT, only 4000 values.

section = {} # Create a dict, split into sections
i=0
jump=10
## Trial debug
while i < 50:
#while i < combined.shape[0]+jump:
    section[int(i/jump)]=combined.iloc[i:i+jump-1]
    print(combined.iloc[i:i+jump-1]['postal'])
    i+=jump
```

```python
print(section.keys())
```

```python
i=0
for key, part in section.items():
    print(f"Current section:{key*jump} to {(key+1)*jump-1}")
    print(f"Total no. of sections: {len(section)}\n")

    #print(part)
    part[['hdbtomrt_distance', 'hdbtomrt_time']]  = part.apply(router, axis=1, result_type='expand')
    writer = pd.ExcelWriter('hdb_to_mrt.xlsx', engine = 'openpyxl',mode='a',if_sheet_exists='replace')
    part.to_excel(writer, sheet_name = f'page{key}')
    writer.close()
```

```python
geylangb[['hdbtomrt_distance', 'hdbtomrt_time']]  = geylangb.apply(walker, axis=1, result_type='expand')
writer = pd.ExcelWriter('hdb_to_mrt - gb.xlsx', engine = 'openpyxl',mode='a',if_sheet_exists='replace')
geylangb.to_excel(writer, sheet_name = 'gb')
writer.close()
```
